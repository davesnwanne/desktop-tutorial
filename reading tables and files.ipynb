# summing a column


```python
#vertical summation of 1 column(top to bottom)

eng["MTH 102"].sum()
```

```python
#total marks obtained in MTH 102 and BIO 101
eng[["MTH 102","BIO 101"]].sum()
```

# summing horizontally (axis = 1)

```python
horsum=eng[["MTH 102","BIO 101","CIT 142","PHY 192"]].sum(axis=1)
# horizontal summation: axis=1
horsum
```

```python
eng["total"]= horsum
eng
```

```python
#calculating the max of the rows

eng[["MTH 102","BIO 101","CIT 142","PHY 192"]].max(axis=1)
maxvar=eng[["MTH 102","BIO 101","CIT 142","PHY 192"]].max(axis=1)
maxvar
```

```python
eng["max"]=maxvar
eng
```

```python
# calculating the average of the rows

eng[["MTH 102","BIO 101","CIT 142","PHY 192"]].mean(axis=1)
avgvar=eng[["MTH 102","BIO 101","CIT 142","PHY 192"]].mean(axis=1)
avgvar
```

```python
eng["average"]=avgvar
eng
```
