# STANDARD SCALER

```python
from sklearn.preprocessing import StandardScaler
x_train= StandardScaler().fit_transform(x_train)
x_test=StandardScaler().fit_transform(x_test)

#standard scaler is used to scale the features/independent variable
#if the values in the independent variables are large,
#the values can be scaled with the standard scaler

#improves performance of scaler

```

```python
model = KNeighborsClassifier(n_neighbors=5).fit(x_train,y_train)
# in knn classifiers you must assign a number N_neighbors=5
```

```python
prediction=model.predict(x_train)
```

```python
# confusion matrix
#show the TP, TN, FP, FN

from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_train,prediction)
```

```python
# accuracy score
# know how well the model performed

from sklearn.metrics import accuracy_score

accuracy_score(y_train, prediction)
```
