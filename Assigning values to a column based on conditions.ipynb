```python
#flour price - 35000
#cashew price - 10000
#cotton price- 6500
#textile price-15000
#rice price- 40000
#beans price-25000
```

```python
#create a user defined function that assigns price to all products
def amount(x):
    if x=="flour":
        save=35000
    elif x=="cashew":
        save=10000
    elif x=="cotton":
        save=6500
    elif x=="textile":
        save=15000
    elif x=="rice":
        save=40000
    elif x=="beans":
        save=25000
    return save
```

# user defined functions are not be quoted eg def amount(x)


```python
#.apply is used to apply a function
mill["Product"].apply(amount)
```

```python
mill["price"]=mill["Product"].apply(amount)
pd.set_option("display.max_rows",None)
mill["price"]
```

```python
#create a function to get grades for score columns
def grade(a):
    if a>=70 and a <100:
        result="A"
    elif a>=60 and a <70:
        result="B"
    elif a>=50 and a <60:
        result="C"
    elif a>=40 and a <50:
        result="D"
    else:
        result="F"
    return result
```

```python
school["MTH 102"].apply(grade)
```

```python
school["grade MTH"]=school["MTH 102"].apply(grade)
school
```
