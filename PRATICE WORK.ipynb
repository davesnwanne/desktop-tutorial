# opening different sheets in a workbook

```python
fruit=pd.read_excel("fruit product data-2.xlsx","Sheet1")
fruit
```

```python
#nunique is usually used on a dataframe or a table
fruit.nunique()
```

```python
#unique is used on a column and shows the unique products in a column
# all unique products in a column
fruit["Product"].unique()
```

```python
#counts the amoount of times a product appears.value_counts
fruit["Product"].value_counts()
```

```python
# total amounts for all products
fruit.groupby("Product").sum().reset_index()
```

```python
total=fruit.groupby("Product").sum().reset_index()
total
```

```python
#sum of amount by state
fruit.groupby("State").sum().reset_index()
```

```python
#nlargest() is used to extract the largest column in the table
#nsmallest() is used to extract smallest column in a table
fruit.groupby("State").sum().nlargest(3,"Amount")
```

```python
fruit.groupby("State").sum().nsmallest(3,"Amount")
```

# arranging the column in an ascending order

```python
sorted(fruit)
```

```python
#arrange the columnns
fruit= fruit[['Amount', 'Category', 'Order Date', 'Order ID', 'Product', 'State']]
fruit
```

```python
#grouping 2 columns
#2 criteria
# when grouping two columns at a time use the ([]) together
fruit.groupby(["Product","State"]).sum()
```

```python
# apple sales in all states breakdown
fruit.groupby(["Product","State"]).sum().loc["Apple"]
```

```python
fruit.groupby(["Product","State"]).sum()["Amount"]
```

```python
#apple amount in ogun state

fruit.groupby(["Product","State"]).sum().loc["Apple"].loc["Ogun"]
```
