```python
x=mental[["Choose your gender","Marital status","Do you have Depression?","Do you have Anxiety?","Do you have Panic attack?","Did you seek any specialist for a treatment?"]]
y=mental["What is your CGPA?"]
```

# split your X and Y

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

# Train and fit the model
# use either logistics or knn modelling for classification
# linear is used for numerical values

```python
model = LogisticRegression().fit(x_train,y_train)
```

```python
model.predict(x_train)
```

```python
y_predict=model.predict(x_train)
```

```python
# checking the output in the dataset and predicted values
```

```python
table=pd.DataFrame(y_train)
# store the x_train in the table variable
```

```python
# create a new column in table
table["prediction"]=y_predict
```

```python
table
```

# NOW PREDICT FOR THE REMAIN PORTION
# WHICH IS THE X_TEST

```python
x_predict=model.predict(x_test)
```

```python
table2=pd.DataFrame(y_test)
# store the x_test in the new variable
```

```python
#create a new column in the table
table2["prediction"]=x_predict
```

```python
table2
```

```python
# bringing the y_train and y_test together
#and the prediction on x_train and x_test
comparison=pd.concat([table, table2],axis=0)
```

```python
comparison= comparison.sort_index()
```
