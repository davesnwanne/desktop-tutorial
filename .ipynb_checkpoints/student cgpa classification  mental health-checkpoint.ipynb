```python
import pandas as pd
```

```python
mental=pd.read_csv("Student Mental health.csv")
```

```python
mental
```

```python
mental.isnull().sum()
#one empty so fill it
```

```python
mental=mental.fillna(0)
#fill the empty cell and assign back to grade
```

```python
mental.dtypes
```

```python
#assign each gender 
# male -1
# female - 0
mental["Choose your gender"].map({"Male":1,"Female":0})
#assign back to the dataframe
mental["Choose Your gender"]=mental["Choose your gender"].map({"Male":1,"Female":0})
```

```python
mental["What is your CGPA?"].unique()
```

```python
#3.00 - 3.49 - 1
#3.50 - 4.00 - 2
#3.50 - 4.00 - 3
#2.50 - 2.99 -4
#2.00 - 2.49 - 5
# 0 - 1.99 - 6
#the numbers in are seen as objects
mental["What is your CGPA?"].map({"3.00 - 3.49":1,"3.50 - 4.00":2,"3.50 - 4.0":3,"2.50 - 2.99":4,"2.00 - 2.49":5,"0 - 1.99":6 })

mental["What is your CGPA?"]=mental["What is your CGPA?"].map({"3.00 - 3.49":1,"3.50 - 4.00":2,"3.50 - 4.0":3,"2.50 - 2.99":4,"2.00 - 2.49":5,"0 - 1.99":6 })

```

```python
mental["Marital status"].astype(str)
mental["Marital status"].map({"NO":0,"YES":1})
mental["Marital status"]=mental["Marital status"].map({"No":0,"Yes":1})
```

```python
mental["Do you have Depression?"].astype(str)
mental["Do you have Depression?"].map({"No":0,"Yes":1})
mental["Do you have Depression?"]=mental["Do you have Depression?"].map({"No":0,"Yes":1})
```

```python
mental["Do you have Anxiety?"].astype(str)
mental["Do you have Anxiety?"].map({"No":0,"Yes":1})
mental["Do you have Anxiety?"]=mental["Do you have Anxiety?"].map({"No":0,"Yes":1})
```

```python
mental["Do you have Panic attack?"].astype(str)
mental["Do you have Panic attack?"].map({"No":0,"Yes":1})
mental["Do you have Panic attack?"]=mental["Do you have Panic attack?"].map({"No":0,"Yes":1})
```

```python
mental["Did you seek any specialist for a treatment?"].astype(str)
mental["Did you seek any specialist for a treatment?"].map({"No":0,"Yes":1})
mental["Did you seek any specialist for a treatment?"]=mental["Did you seek any specialist for a treatment?"].map({"No":0,"Yes":1})
```
