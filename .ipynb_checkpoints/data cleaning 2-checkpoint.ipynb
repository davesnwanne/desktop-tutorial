```python
import pandas as pd
usa=pd.read_excel("Data Cleaning Excel Tutorial-1.xlsx")
```

```python
#deleting a column
del usa["Unnamed: 0"]
```

```python
#show all the contents in the column i.e increase column width
pd.set_option("display.max_colwidth",None)
```

```python
usa.dtypes
```

```python
 #removing errors using regex df["columname to be fixed"].srt(because regex works on with strings).(\W','',regex=true)
usa["prior"].str.replace("\W"," ",regex=True)
```

#when you want to remove errors at once in a dataframe use keyword(for)
# for loop. for in
#for loop is a function that makes work faster in coding .

```python
for clear in usa.columns:
    print(clear)
```

```python
for clear in usa.columns[1:5]:
    usa[clear]=usa[clear].str.replace("United States", "US")
    usa[clear]=usa[clear].str.replace("\W", " ", regex=True)
```
