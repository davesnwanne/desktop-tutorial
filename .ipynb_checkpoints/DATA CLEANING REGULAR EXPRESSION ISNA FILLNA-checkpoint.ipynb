```python
import pandas as pd
vaccine=pd.read_csv("covid vaccine.csv")
```

```python
vaccine.dtypes
```

```python
vaccine["Total population"]=vaccine["Total population"].str.replace(",","")
```

```python
vaccine["Total Doses Administered (Astra)"]=vaccine["Total Doses Administered (Astra)"].str.replace(","," ")
```

```python
vaccine["Total Doses Administered (Moderna)"]=vaccine["Total Doses Administered (Moderna)"].str.replace(","," ")
```

# Regular expressions ("\W")

```python
#we use regular expressions to remove errors in a column  or data frame
# we pass into the .sr.replace() function

#regular expressions can remove characters not numbers nor text
```

```python
vaccine["State"]=vaccine["State"].str.replace("\W","",regex=True)
vaccine["State"]
```
