```python
def grade(y):
    if y>=70 and y <100:
        result ="A"
    elif y>=60 and y <70:
        result="B"
    elif y>=50 and y <60:
        result="C"
    elif y>40 and y <50:
        result="D"
    elif y>=30 and y <40:
        result ="E"
    else:
        result="F"
    return result
        
```

```python
performance["writing score"].apply(grade)
```

```python
performance["gradewriting"]=performance["writing score"].apply(grade)
```
